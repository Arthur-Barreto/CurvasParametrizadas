### •  Pontos em que a derivada é zero


```python
intervalo = np.arange(0, 6*math.pi, 0.00001)

# negativo = 0
# positivo = 1
```

##### Pontos em que a reta tangente é vertical

```python
def fx(t):
    return 2*math.sin(4*t/3) - math.sin(t)

sinais_x = list()
raizes_x = list()

for raiz in intervalo:
    # Checando se o valor é positivo
    if fx(raiz) > 0:
        sinais_x.append(1)
    else:
        sinais_x.append(0)
        
i = 1
while i < len(sinais_x):
    if sinais_x[i] != sinais_x[i-1]:
        raizes_x.append(intervalo[i])
        
    i+=1
        
raizes_x
```

##### Pontos em que a reta tangente é horizontal

```python
def fy(t):
    return 4*math.cos(t) - 8*math.cos(4*t/3)

sinais_y = list()
raizes_y = list()

for raiz in intervalo:
    # Checando se o valor é positivo
    if fy(raiz) > 0:
        sinais_y.append(1)
    else:
        sinais_y.append(0)
        
i = 1
while i < len(sinais_y):
    if sinais_y[i] != sinais_y[i-1]:
        raizes_y.append(intervalo[i])
        
    i+=1
        
raizes_y
```

## •  Comprimento

```python
def x(t):
    return 4*math.cos(t) - 6*math.cos(4*t/3)
def y(t):
    return 4*math.sin(t) - 6*math.sin(4*t/3)
```

##### Aproximação pela poligonal

```python
parametros = np.arange(0, 6*math.pi, 6*math.pi/10000)


def distanciaEuclidiana(A,B):
    return ( (A[0] - B[0])**2 + (A[1] - B[1])**2 )**0.5

pontos = list()

for i in parametros:
    ponto = (x(i),y(i))
    pontos.append(ponto)
    
comprimento = 0
c = 1
while c < len(pontos):
    d = distanciaEuclidiana(pontos[c],pontos[c-1])
    comprimento += d
    
    c+=1
    
comprimento
```

##### Aproximação por figuras mais simples

```python
def x(t):
    return 4*math.cos(t) - 6*math.cos(4*t/3)
def y(t):
    return 4*math.sin(t) - 6*math.sin(4*t/3)
```

```python
intervalo = np.arange(v3, u3, (u3-v3)/1000)

def areaTriangulo(A,B):
    determinante = abs(A[0]*B[1] - A[1]*B[0])
    return 0.5*determinante

pontos = list()

for t in intervalo:
    pontos.append([x(t),y(t)])
    
area = 0
i = 1
while i < len(pontos):
    area += areaTriangulo(pontos[i],pontos[i-1])
    i+=1
    
area
```
